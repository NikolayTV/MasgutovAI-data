```python
import os

def add_txt_extension(directory):
    for root, dirs, files in os.walk(directory):
        for filename in files:
            if '.' not in filename:  # Check if filename has no extension
                old_path = os.path.join(root, filename)
                new_path = old_path + '.txt'
                os.rename(old_path, new_path)
                print(f'{old_path} renamed to {new_path}')

# Example usage:
# Specify the directory path you want to iterate over recursively
directory_path = '../content'
add_txt_extension(directory_path)
```

```python
import os, json

def dir_to_json_with_txt_content(path):
    """
    Recursively walks through a directory path, creating a nested dictionary structure that mirrors the
    directory structure. For .txt files, includes their content as the value.
    """
    # The structure to represent either a directory or the content of a .txt file
    structure = {}
    for item in os.listdir(path):
        item_path = os.path.join(path, item)
        if os.path.isdir(item_path):
            # Recurse into subdirectories
            structure[item] = dir_to_json_with_txt_content(item_path)
        # elif item.endswith('.txt'):
        else:
            # Read the content of .txt files
            with open(item_path, 'r', encoding='utf-8') as file:
                structure[item] = file.read()
    return structure

# Specify the root directory path you wish to convert to JSON
root_directory_path = '../content'
nested_json = dir_to_json_with_txt_content(root_directory_path)
with open('../doc.json', 'w', encoding='utf-8') as file:
    json.dump(nested_json, file, ensure_ascii=False, indent=4)

```
